## Research Question

How well do students' sleep habits, including insomnia symptoms, self-reported sleep quality, and nightly sleep duration, predict their academic performance measures such as focus, motivation, and grades?

## Background and Prior Work

Sleep is a periodically recurring state of rest that occurs in the human body every 24 hours and lasts for several hours. It is naturally induced by the brain and is characterized by reduced activity and decreased responsiveness to stimuli. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The functions of sleep in the brain include clearing out toxins, consolidating information learned during the day, and repairing neurons damaged by free radicals. These processes support the cognitive systems required for effective learning, attention, and decision-making. Because academic performance depends heavily on these same cognitive functions, it is reasonable to expect that sleep quality could predict how well students are able to perform academically. This connection forms the basis of our project.  

There are numerous existing studies that demonstrate how sleep has a significant influence on cognitive functioning. For example, a study by Garcia et al. found that all basic cognitive processes of people were detrimentally affected after 24 hours or more without sleep, including attention, working memory, and executive functions such as cognitive flexibility and inhibition of irrelevant sensory stimuli. However, the researchers also found that these processes were differentially affected by sleep deprivation. Attention related processes were found to be impacted the most, while working memory was impacted moderately and cognitive flexibility, one component of executive functioning, was not found to be impacted at all. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) These researchers were able to quantify the effects of sleep deprivation through cognitive tests given before and after the participants were subjected to the sleep deprivation. These findings are especially relevant for our study because academic tasks such as focusing in class, completing assignments, and maintaining motivation depend heavily on attention and working memory, the processes most disrupted by poor sleep quality.

Narrowing down further, researchers from the University of Washington examined sleep and academic performance using measures such as chronotype, sleep variability, and differences between weekday and weekend sleep timing <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). They found that irregular or low quality sleep predicted worse academic performance even when total sleep duration was similar across students. This suggests that subjective sleep quality indicators such as insomnia symptoms may capture important differences in cognitive and academic functioning.

Across these studies, a clear pattern appears: when sleep is disrupted, whether through lack of sleep, irregular schedules, or ongoing sleep problems, attention and working memory tend to decline. These abilities are essential for doing well in school. Our study builds on this work by focusing specifically on insomnia symptoms as our measure of sleep quality and students’ self-reported academic behaviors as our measure of academic performance. Insomnia symptoms capture difficulties falling asleep, staying asleep, or experiencing restorative rest, all of which are known to interfere with cognitive performance. Likewise, academic behaviors such as focus, motivation, and assignment completion offer a practical and behaviorally grounded way to assess how these cognitive changes manifest in students' daily academic lives. By aligning our measurements with the specific processes identified in prior research, our study aims to test whether the same relationships hold within our own sample of students.

## Hypothesis


We expect that students with healthier sleep habits, including better sleep quality, fewer insomnia symptoms, and longer sleep duration, will show stronger academic performance signals in terms of concentration, fatigue, and grades. In particular, students who report more severe insomnia symptoms or poorer sleep quality are likely to report lower levels of focus, motivation, and grade averages.

This prediction follows from prior research showing that poor or insufficient sleep weakens attention, working memory, and other cognitive processes that are essential for academic success.

### Data overview

### Dataset #1: Student Insomnia and Educational Outcomes
This dataset focuses on understanding the relationship between students’ sleep habits, lifestyle factors, and academic performance. Each row represents an individual survey response, and each column corresponds to a self-reported variable related to sleep behavior, stress, and performance. Although only two sample responses are shown, the full dataset reportedly includes hundreds of participants, providing insight into how insufficient or poor-quality sleep affects students’ academic and cognitive functioning. Since there are two versions of the dataset, we will be sticking with Version 2, which has 996 rows of data.

#### Data Collection
The data was collected through an online survey administered via Google Forms in Oct-Nov 2024. Respondents were asked to provide insights into their sleep behaviors and the effects on their academic and daily activities.

#### Key Features
1. Demographics: Year of study and gender.
2. Sleep Patterns: Frequency of difficulty falling asleep, hours of sleep, night awakenings, and overall sleep quality.
3. Cognitive and Academic Effects: Impact on concentration, fatigue, class attendance, assignment completion, and overall academic performance.
4. Lifestyle Factors: Electronic device usage before sleep, caffeine consumption, and physical activity frequency.
5. Stress Levels: Self-reported stress related to academic workload.

#### Concerns
Since all data are self-reported, responses may be influenced by the participants ability to recall information or an effort to make themselves look better. For example, students may overestimate how much sleep or exercise they get, or underreport screen time and caffeine consumption. Moreover, because the survey was optional, students that filled it out were likely more aware or concerned about their sleep habits. This could lead to potential sampling bias. Measures such as “sleep quality”, “stress”, and “academic performance” are also subjective, leading to a lack of standardizationg and comparability of results across individuals. Another limitation lies in the timing and context of data collection. The survey was conducted between October and November 2024, when midterm or final exams often occur. As a result, there would be more academic stress and worse sleep quality during this period, meaning the responses may not accurately represent students’ typical sleep behaviors throughout the semester.

Despite these limitations, the dataset remains valuable for exploring general trends and relationships between sleep, lifestyle factors, and academic performance, offering meaningful insights into how sleep hygiene affects university students’ well-being and success.

### Dataset #2: Student Performance & Behavior Dataset
This dataset is a substantial real-world collection of 5,000 records gathered by a private learning provider. It offers a multifaceted view of student success by integrating academic performance metrics with behavioral and socio-economic context. The dataset is ideal for exploring patterns, correlations, and predictive insights related to academic achievement.

#### Data Collection
The data was collected from 5,000 actual student records managed by a private learning provider. It includes information captured across the duration of a course or academic period, providing a comprehensive set of variables necessary for exploring the factors that influence academic outcomes.

#### Key Features
1. Academic Outcomes: The primary dependent variables are the calculated Total_Score and the assigned letter Grade (A, B, C, D, F).
2. Academic Inputs: Detailed performance scores, including Midterm_Score, Final_Score, Assignments_Avg, Quizzes_Avg, Participation_Score, and Projects_Score. Total Score Calculation: The Total_Score is a weighted sum. The Projects Score (30%) and Final Score (25%) carry the most weight, while Attendance (%) has minimal or no influence.
3. Behavioral & Lifestyle Factors: Variables capturing student habits and resources, such as Study_Hours_per_Week, Sleep_Hours_per_Night, Stress_Level (1-10), Extracurricular_Activities, and Internet_Access_at_Home.
4. Demographics & Socio-economic: Identifying and grouping features like Student_ID, Gender, Age, Department, Parent_Education_Level, and Family_Income_Level.

#### Concerns
As real-world data, this dataset presents several limitations and caveats essential for proper analysis:
1. Missing Values (Nulls): The dataset is not complete; records contain missing values (nulls) in some critical areas, such as Attendance (%), Assignments_Avg, and Parent_Education_Level. Analysts must decide on appropriate imputation or handling strategies.
2. Data Bias: There is an acknowledged bias in grading where factors not explicitly weighted—specifically, high attendance—can lead to slightly better final grades. This suggests the final letter Grade is not solely determined by the calculated Total_Score and may be influenced by subjective instructor practices.
3. Imbalanced Distributions: The student population is not uniform across all categories. There are imbalanced distributions across departments, meaning some fields (e.g., CS) may have significantly more students than others (e.g., Business). This skew should be considered when conducting comparative analyses.
4. Subjectivity: Variables like Stress_Level (1-10) are self-reported and inherently subjective, lacking clinical standardization. The Participation_Score is also subject to instructor judgment, introducing variability.

### additional analysis - calculating the p-values for the correlation between sleep variables and academic variables in both Dataset 1 and Dataset 2

```python
df1 = df_1
df2 = df_2

# -------------------------------------------------
# Convert all numeric-looking columns to numeric
# -------------------------------------------------
for df in [df1, df2]:
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# -------------------------------------------------
# DEFINE sleep and academic variables correctly
# -------------------------------------------------

sleep_vars_1 = [
    "sleep_hours",
    "sleep_quality",
    "diff_fall_asleep",
    "wake_during_night",
    "sleep_impact_assignments"
]

acad_vars_1 = [
    "concentration_issues",
    "daytime_fatigue",
    "miss_classes",
    "academic_stress",
    "academic_perf"
]

sleep_vars_2 = ["sleep_hours_per_night"]

acad_vars_2 = [
    "total_score",
    "grade",
    "study_hours_per_week",
    "extracurricular_activities",
    "stress_level"
]

# -------------------------------------------------
# Compute correlations for Dataset 1
# -------------------------------------------------
rows1 = []
for s in sleep_vars_1:
    for a in acad_vars_1:
        if s in df1.columns and a in df1.columns:
            valid = df1[[s, a]].dropna()
            if len(valid) > 1:
                r, p = pearsonr(valid[s], valid[a])
                rows1.append({
                    "sleep_var": s,
                    "acad_var": a,
                    "p_value": p
                })

sleep_pvals1 = pd.DataFrame(rows1)
sleep_pvals1["label"] = sleep_pvals1["sleep_var"] + " vs " + sleep_pvals1["acad_var"]

# -------------------------------------------------
# Compute correlations for Dataset 2
# -------------------------------------------------
rows2 = []
for s in sleep_vars_2:
    for a in acad_vars_2:
        if s in df2.columns and a in df2.columns:
            valid = df2[[s, a]].dropna()
            if len(valid) > 1:
                r, p = pearsonr(valid[s], valid[a])
                rows2.append({
                    "sleep_var": s,
                    "acad_var": a,
                    "p_value": p
                })

sleep_pvals2 = pd.DataFrame(rows2)
sleep_pvals2["label"] = sleep_pvals2["sleep_var"] + " vs " + sleep_pvals2["acad_var"]

# -------------------------------------------------
# SORT BY p-value for readability
# -------------------------------------------------
sleep_pvals1 = sleep_pvals1.sort_values("p_value")
sleep_pvals2 = sleep_pvals2.sort_values("p_value")

# -------------------------------------------------
# Bar Chart — Dataset 1
# -------------------------------------------------
plt.figure(figsize=(12, 10))
plt.barh(sleep_pvals1["label"], sleep_pvals1["p_value"])
plt.axvline(0.05, color="black", linestyle="--", linewidth=1)
plt.xlabel("p-value")
plt.title("Dataset 1: P-values for Sleep vs Academic Performance")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.xscale("log")
plt.show()

# -------------------------------------------------
# Bar Chart — Dataset 2
# -------------------------------------------------
plt.figure(figsize=(12, 6))
plt.barh(sleep_pvals2["label"], sleep_pvals2["p_value"])
plt.axvline(0.05, color="black", linestyle="--", linewidth=1)
plt.xlabel("p-value")
plt.title("Dataset 2: P-values for Sleep vs Academic Performance")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

p-value analysis: The chart for Dataset 1 shows that all correlations between subjective sleep deficits (like sleep_quality or diff_fall_asleep) and academic difficulties (concentration_issues, daytime_fatigue) are highly statistically significant (p < 0.05), strongly supporting the initial prediction. Conversely, the chart for Dataset 2 shows that correlations between objective sleep_hours_per_night and final grades (total_score, grade) are not statistically significant (p > 0.05). This critical comparison concludes that while insufficient or poor sleep quality significantly impairs the cognitive functions required for daily academic tasks, the simple quantity of sleep hours has no measurable, statistically reliable predictive power over a student's final, weighted course scores.
